```python
import tensorflow as tf
from imutils import paths
import matplotlib.pyplot as plt
import numpy as np
import cv2
import os
import random
```

```python
imagePaths = list(paths.list_images("Dataset/IMFDB_final"))
```

```python
print(len(imagePaths))
```

```python
names = []
for imagePath in imagePaths:
    names.append(imagePath.split("/")[-4])
print(set(names))
```

```python
os.mkdir("Dataset/IMFDBWF")
```

```python
names = set(names)
for name in names:
    string = name
    path = "Dataset/IMFDBWF/"+str(name)
    print(path)
    os.mkdir(path)
    for (i,imagePath) in enumerate(imagePaths):
        if name == imagePath.split("/")[-4]:
            img = cv2.imread(imagePath)
            img_path = path+"/"+str(i)+".jpg"
            print(img_path)
            cv2.imwrite(img_path , img)
```

```python
for name in names:
    path = "Dataset/IMFDBWF/"
    write_path = "Dataset/IMFDBTT/train/"+str(name)
    os.mkdir(write_path)
    pathimages = list(paths.list_images(path))
    size = len(pathimages)
    train_size = int(0.7 * size)
    test_size = size - train_size
    print("Train Size for",name,train_size)
    print("Test Size for",name, test_size)
    a = random.sample(range(0,size),size)
    for i in range(0,size):
        if i < train_size:
            path_name = write_path+"/"+str(i)+".jpg"
            img = cv2.imread(pathimages[a[i]])
            cv2.imwrite(path_name,img)
            
```

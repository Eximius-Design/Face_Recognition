```python
import tensorflow as tf
import cv2
from imutils import paths
import numpy as np
import pickle
from src import Detection
from src import Recognition
from src import preprocess
import matplotlib.pyplot as plt
import math
import imutils
```

```python
imagePaths = list(paths.list_images("/home/santoshv/Dataset/test/"))
print(imagePaths)
```

```python
imagePaths[0].split("/")[-2]
```

```python
with open("util/EximiusAI_H.pkl", 'rb') as infile:
    model,class_names = pickle.load(infile)
```

```python
rec = Recognition.Recognition("Models/Recognition_facenet.pb")
```

```python
correct = 0
wrong = 0
for imagePath in imagePaths:
    image = cv2.imread(imagePath)
    image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
    name = imagePath.split("/")[-2]
    print("Original Name:",name)
    image = preprocess.prewhiten(image)
    emb = rec.recognize(image)
    predicted_name = model.predict(emb)
    print("Predicted Name:",predicted_name)
    if predicted_name == name:
        correct = correct + 1
    if predicted_name != name:
        wrong = wrong + 1
    print("corrects:",correct)
    print("wrongs:",wrong)
```

```python
accuracy = correct/(correct+wrong)
print(accuracy*100)
```

```python
def trans(x,y,a):
    A = (np.array(img.shape[0:2])/2)[1]
    B = (np.array(img.shape[0:2])/2)[0]
    print(str(x)+" = "+str(x)+" - "+str(A))
    print(str(y)+" = "+str(y)+" + "+str(B))
    x = x - A
    y = y + B
    print(x, y , A, B)
    X = x * math.cos(a) - y * math.sin(a)
    Y = x * math.sin(a) + y * math.cos(a)
    X = X.astype('int32')
    Y = Y.astype('int32')
    return (X, Y) 
```

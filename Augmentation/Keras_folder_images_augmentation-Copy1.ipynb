```python
from keras.preprocessing.image import ImageDataGenerator
import os 
from imutils import paths
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
input_path = "../Dataset/AI_team/tmp/"
output_path = "../Dataset/AI_team_augmented/tmp/"
batch_size = 30
```

```python
image_gen = ImageDataGenerator(rotation_range=15,
                               width_shift_range=0.1,
                               height_shift_range=0.1,
                               shear_range=0.01,
                               zoom_range=[0.9, 1.25],
                               horizontal_flip=True,
                               vertical_flip=False,
                               data_format='channels_last',
                               brightness_range=[0.5, 1.5])
```

```python
imagePaths = list(paths.list_images(input_path))
print(len(imagePaths))
```

```python
count = 0
for i in imagePaths:
    count = count + 1
    print("[INFO] AUGMENTING IMAGE...", count)
    imagepath = i
    print(imagepath)
    image = np.expand_dims(cv2.cvtColor(cv2.imread(imagepath),cv2.COLOR_BGR2RGB),0)
    aug_iter = image_gen.flow(image, batch_size=30)
    aug_images = [next(aug_iter)[0].astype(np.uint8) for i in range(batch_size)]
    copy = 0
    for j in aug_images:
        j = cv2.cvtColor(j,cv2.COLOR_RGB2BGR)
        copy = copy + 1
        cv2.imwrite(output_path+str(count)+"_"+str(copy)+".jpg",j)
        print("[INFO] WRITING IMAGE COPY", copy)
```
